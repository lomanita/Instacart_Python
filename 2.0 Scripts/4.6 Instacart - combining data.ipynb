# 3.0 Data wrangling and merging

```python
# Dropping the _merge column from df_orders_and_products

df_orders_and_products = df_orders_and_products.drop(columns = '_merge')
```

```python
# Merging the df_orders_and_products and df_products

df_orders_products_merged = df_orders_and_products.merge(df_products, on = 'product_id', indicator = True)
```

```python
# Checking the result of the merging

df_orders_products_merged['_merge'].value_counts()
```

```python
# Exporting the df_orders_products_merged as pickle

df_orders_products_merged.to_pickle(os.path.join(path, '2.2 Prepared data', 'orders_products_merged.pkl'))
```
